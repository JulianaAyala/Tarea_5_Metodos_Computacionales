```python
%pylab inline
from scipy.fftpack import fft, fftfreq
```

```python
#Señal de prueba
#Número de puntos
n=100  
#Frecuencia en Hz
f=400.0 
#Espaciado, 16 puntos por período
dt=1/(f*16)
#Intervalo de tiempo en segundos
t=linspace(0,(n-1)*dt,n)  
#Señal
y=sin(2*pi*f*t)-(1/2)*sin(2*pi*2*f*t)
 
plot(t,y)
scatter(t,y,c='r')
xlabel('Tiempo (s)')
ylabel('y(t)')
```

```python
#Tranformada de Fourier
Y=fft(y)
#Arreglo con la frecuencia de cada punto de la señal
frecuencias=fftfreq(n,dt)
#Arreglo que va a guardar los puntos filtrados en el dominio de la frecuencia
Y_filtrada=zeros(n)
#Definición de la banda de frecuencias a filtrar
frec_min=100
frec_max=200
for i in range(n):
    #Si la frecuncia del punto no está dentro de la banda especificada, su valor es cero
    if(frecuencias[i]<frec_min or frecuencias[i]>frec_max):
        Y_filtrada[i]=0
    else:
        Y_filtrada[i]=Y[i]
#Transformada inversa de Fourier
y_filtrada=ifft(Y_filtrada)
plot(t,y_filtrada)
#scatter(t,y_filtrada,c='r')
plot(t,y,c='g')
#scatter(t,y,c='k')
xlabel('Tiempo (s)')
ylabel('y(t)')
```
